3. Какой магазин заработал больше всех за период с 2012 по 2013 гг.?

```
table.loc[:, ('Стоимость. руб.', 2012):('Стоимость. руб.', 2013)]
```

```
df[(df['Дата'].dt.year >= 2012) & (df['Дата'].dt.year <= 2013)]
```

```
table=pd.pivot_table(df[(df['Дата'].dt.year >= 2012) & (df['Дата'].dt.year <= 2014)],
               index=["Магазин"],
               values=["Стоимость. руб."],
               aggfunc=np.sum,
               fill_value=0)              
table.sort_values(by='Стоимость. руб.', ascending=False)
```

4. Какой % занимают овощи и фрукты в структуре продаж магазинов г. Владимир?

```
df[df['Город'] == 'Владимир']
```

```
table=pd.pivot_table(df[df['Город'] == 'Владимир'],
               index=["Категория"],
               values=["Стоимость. руб."],
               aggfunc=np.sum,
               fill_value=0,
              margins=True)
table.sort_values(by='Стоимость. руб.', ascending=False)
```

```
table=pd.pivot_table(df[df['Город'] == 'Владимир'],
               index=["Товар"],
               values=["Стоимость. руб."],
               aggfunc=np.sum,
               fill_value=0,
              margins=False)
table.sort_values(by='Стоимость. руб.', ascending=False)
```

```
table['% of Total'] = (table['Стоимость. руб.'] / table['Стоимость. руб.'].sum() * 100)
```
