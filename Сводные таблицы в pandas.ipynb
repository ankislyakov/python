```
import pandas as pd
import numpy as np
```

```
# берем данные любые с https://data-in.ru/data-catalog/
df = pd.read_csv("/content/drive/MyDrive/фруктыовощи (2).csv", sep=';')
df.head()




```

```
df["Магазин"] = df["Магазин"].astype("category")
df["Город"] = df["Город"].astype("category")
df["Товар"] = df["Товар"].astype("category")

```

```
df.info()
```

```
# по умолчанию усредняются
pd.pivot_table(df,
               index=["Магазин"])
```

```
pd.pivot_table(df,
               index=["Магазин", "Город","Категория"])
```

```
pd.pivot_table(df,
               index=["Категория","Магазин", "Город"])
```

```
pd.pivot_table(df,
               index=["Категория","Магазин", "Город"],
               values=["Стоимость. руб."])
```

```
pd.pivot_table(df,
               index=["Категория","Магазин", "Город"],
               values=["Стоимость. руб."],
               aggfunc=np.sum)
```

```
pd.pivot_table(df,
               index=["Категория","Магазин", "Город"],
               values=["Стоимость. руб."],
                aggfunc=[np.mean, len])
```

```
pd.pivot_table(df,
               index=["Категория", ],
               columns=["Город"],
               values=["Стоимость. руб."],
               aggfunc=np.sum,
              fill_value=0,
               margins=True)
```

```
pd.pivot_table(df,
               index=["Категория", "Товар"],
               columns=["Город"],
               values=["Стоимость. руб."],
               aggfunc=np.sum,
              fill_value=0,
               margins=True)
```

```
table=pd.pivot_table(df,
               index=["Категория", "Товар"],
               columns=["Город"],
               values=["Стоимость. руб.", "Количество. кг"],
               aggfunc={"Количество. кг": sum, "Стоимость. руб.": np.sum},
              fill_value=0)


```

```
table
```

```
table.query('Товар == ["Баклажаны","Яблоки"]')
```

```
table.query('Товар == ["Баклажаны","Яблоки"] and Категория == ["Овощи"]')
```
